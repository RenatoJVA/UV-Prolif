```python

import ipywidgets as widgets
from IPython.display import display
import warnings
warnings.filterwarnings("ignore")
button = widgets.Button(description="Load imports", button_style="info")
output = widgets.Output()

def load_imports(b):
    with output:
        output.clear_output()
        print("Importing libraries... ", end="")

        # ===========================
        # Standard library
        # ===========================
        import os, sys

        # ===========================
        # Scientific & data handling
        # ===========================
        import numpy as np, pandas as pd

        # ===========================
        # Molecular analysis
        # ===========================
        import mdtraj as md
        import MDAnalysis as mda
        from MDAnalysis.topology.guessers import guess_types
        import prolif as plf
        from prolif.plotting.complex3d import Complex3D

        # ===========================
        # Cheminformatics
        # ===========================
        from rdkit import Chem, DataStructs

        # ===========================
        # Plotting & visualization
        # ===========================
        import matplotlib as mpl
        from matplotlib import pyplot as plt
        import matplotlib.font_manager as fm
        import seaborn as sns

        # ===========================
        # Widgets (Jupyter UI)
        # ===========================
        import ipywidgets as widgets
        from IPython.display import display

        # ===========================
        # Configuración global
        # ===========================
        mpl.rcParams['font.family'] = 'Times New Roman'
        print("done")

button.on_click(load_imports)
display(button, output)
```

```python
# Widgets de entrada
dynamic_path = widgets.Text(value="", description="Ruta:")
ligand_widget = widgets.Text(value="", description="Ligando:")
protein_widget = widgets.Text(value="", description="Proteína:")

# Botón para cargar
load_button = widgets.Button(description="Cargar sistema")

# Área de salida
output_area = widgets.Output()

# Mostrar en pantalla
display(dynamic_path, ligand_widget, protein_widget, load_button, output_area)
```

```python
def update_params():
    global PROJECT_PATH, LIGAND_NAME, PROTEIN_NAME, TPR_FILE, TRAJ_FILE
    PROJECT_PATH = dynamic_path.value.strip()
    LIGAND_NAME = ligand_widget.value.strip()
    PROTEIN_NAME = protein_widget.value.strip()
    
    # Construir paths absolutos
    TPR_FILE = os.path.join(PROJECT_PATH, f"sdm-{LIGAND_NAME}-{PROTEIN_NAME}.tpr")
    TRAJ_FILE = os.path.join(PROJECT_PATH, f"sdm-{LIGAND_NAME}-{PROTEIN_NAME}-noPBC.xtc")

def load_system(_):
    global u, ligand_selection, protein_selection 
    with output_area:
        output_area.clear_output()
        update_params()

        # Validación
        if not os.path.isdir(PROJECT_PATH):
            print(f"La ruta no existe: {PROJECT_PATH}")
            return
        if not os.path.exists(TPR_FILE):
            print(f"No se encontró el archivo: {TPR_FILE}")
            return
        if not os.path.exists(TRAJ_FILE):
            print(f"No se encontró el archivo: {TRAJ_FILE}")
            return

        # Cargar universo de MDAnalysis
        u = mda.Universe(TPR_FILE, TRAJ_FILE)
        ligand_selection = u.select_atoms(f"resname {LIGAND_NAME}")
        protein_selection = u.select_atoms("protein")

        print("Sistema cargado correctamente")
        print(f"- Ruta: {PROJECT_PATH}")
        print(f"- Ligando: {LIGAND_NAME} ({ligand_selection.n_atoms} átomos)")
        print(f"- Proteína: {PROTEIN_NAME} ({protein_selection.n_atoms} átomos)")

# Conectar botón
load_button.on_click(load_system)
```

```python
protein_selection = u.select_atoms(
    "protein and byres around 20.0 group ligand", ligand=ligand_selection
)
protein_selection
```

```python
# Asegurarse de que los atributos 'elements' y 'bonds' estén presentes
if not hasattr(u, "elements"):
    elements = mda.topology.guessers.guess_types(u.atoms.names)
    u.add_TopologyAttr("elements", elements)

if not hasattr(u, "bonds"):
    u.add_TopologyAttr("bonds")

# Convertir la selección a un objeto Molecule de ProLIF, permitiendo hidrógenos implícitos y forzando la conversión
ligand_mol = plf.Molecule.from_mda(ligand_selection, NoImplicit=False, force=True)

# Visualizar los residuos del ligando
plf.display_residues(ligand_mol, size=(400, 200))
```

```python
plf.Fingerprint.list_available()
```

```python
# use default interactions
fp = plf.Fingerprint()
# run on a slice of the trajectory frames: from begining to end with a step of 10
fp.run(u.trajectory[::10], ligand_selection, protein_selection)
```

```python
fp.to_pickle("fingerprint.pkl")
fp = plf.Fingerprint.from_pickle("fingerprint.pkl")
```

```python
df = fp.to_dataframe()
# show only the 10 first frames
df.head(5)
```

```python
# Tanimoto similarity between the first frame and the rest
bitvectors = fp.to_bitvectors()
tanimoto_sims = DataStructs.BulkTanimotoSimilarity(bitvectors[0], bitvectors)
#tanimoto_sims
```

```python
# Tanimoto similarity matrix
bitvectors = fp.to_bitvectors()
similarity_matrix = []
for bv in bitvectors:
    similarity_matrix.append(DataStructs.BulkTanimotoSimilarity(bv, bitvectors))
similarity_matrix = pd.DataFrame(similarity_matrix, index=df.index, columns=df.index)

# display heatmap
fig, ax = plt.subplots(figsize=(5, 5), dpi=200)
colormap = sns.diverging_palette(
    300, 145, s=90, l=80, sep=30, center="dark", as_cmap=True
)
sns.heatmap(
    similarity_matrix,
    ax=ax,
    square=True,
    cmap=colormap,
    vmin=0,
    vmax=1,
    center=0.5,
    xticklabels=5,
    yticklabels=5,
)
ax.invert_yaxis()
plt.yticks(rotation="horizontal")
fig.patch.set_facecolor("white")
```
